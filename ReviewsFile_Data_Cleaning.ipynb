```python
import nltk
import pandas as pd
import string
import re
from nltk.corpus import stopwords

reviewsRawData = pd.read_csv('reviews.csv')
```

```python
comments = pd.DataFrame(reviewsRawData['comments'])
comments.columns = ['comments']

comments.head()
```

```python
#Checking how many of the individual data is not reviewed in the DataSet 
print("Number of nulls in the label: {}".format(reviewsRawData['comments'].isnull().sum()))


#Checking total number of data is available in the column
print("Input data has {} rows and {} columns".format(len(comments), len(comments.columns)))
```

```python
comments['comments'] = comments['comments'].str.replace(r'\d+', '')
text = comments["comments"]


def text_process(text):
    nopunc = [char for char in text if char not in string.punctuation]

    nopunc = ''.join(nopunc)

    return [word for word in nopunc.split() if word.lower() not in
            stopwords.words('english')]


print("RESULTS:")
for row in text:
    print(text_process(row))



#Hi the below code i tried to remove punctuation but there is a Type Error: 'float' object is not iterable.

# def remove_punct(text):
#     text_nopunct = "".join([char for char in text if char not in string.punctuation])
#     return text_nopunct

# comments['body_text_clean'] = comments['comments'].apply(lambda x: remove_punct(x))

# comments.head()
```
